```python
#to find the data dump into given url (flipkart)

big_boxes = flipkart_html.findAll ("div", {"class": "_1AtVbE col-12-12"})
```

```python
#to check length of big boxes
len(big_boxes)
```

```python
#to go into hyperlink of any of boxes --> should check the div section of the data

small_box = big_boxes[2]
```

```python
#to get the url in div section 
search_url = small_box.div.div.div.a["href"]
```

```python
#appending with flipkart_url

product_link = "https://www.flipkart.com" + search_url
```

```python
#to get into product request 
product_req = requests.get(product_link)
```

```python
product_req.text
```

```python
#to beautify more

product_html = bs(product_req.text, "html.parser")
```

```python
#to extract reviews part

comment_box = product_html.find_all("div", {"class" : "_16PBlm"})
```

```python
#to know the length

len(comment_box)
```

```python
#to go into rating secton --> getting into div section for the rating 

comment_box[1].div.div.div.div.text
```

```python
#now looping & obtainig rating of every section

for i in comment_box:
    print(i.div.div.div.div.text)
```
